## GetNetwork

```python
from twitter import GetDescriptionNetwork
```

```python
from twitter import CleansingTweets
from tokenizer import Tokenizer

gn = GetDescriptionNetwork(cleansing=CleansingTweets().cleansing_text, tokenizer=Tokenizer().mecab_tokenizer,
                           max_followers=50)
adj_list, users_info = gn.get_network('Deepblue_ts')
```

```python
import pandas as pd
import numpy as np
df = pd.DataFrame.from_dict(users_info, orient='index').reset_index(drop=True)
```

```python
df['favourites_par_day'] = df[['favourites_count', 'elapsed_date']].apply(lambda x: x[0]/x[1], axis=1)
df['tweet_per_day'] = df[['statuses_count', 'elapsed_date']].apply(lambda x: x[0]/x[1], axis=1)
df.head()
```

```python
df.favourites_par_day.hist(bins=100)
```

```python
df.tweet_per_day.hist(bins=100)
```

```python
np.percentile(df.favourites_par_day, q=[25, 75])
```

```python
np.percentile(df.tweet_per_day, q=[25, 75])
```

```python
import networkx as nx
import matplotlib.pyplot as plt

G = nx.DiGraph()
G.add_weighted_edges_from(adj_list)

plt.figure(figsize=(20, 10))
plt.axis('off')
pos = nx.spring_layout(G, k=0.9)
nx.draw_networkx_nodes(G, pos, node_color="w", alpha=0.6, node_size=1, edgecolors="k")
nx.draw_networkx_labels(G, pos, font_size=10, font_weight="bold")
nx.draw_networkx_edges(G, pos, alpha=0.4, edge_color="c", width=1)
plt.show()
```
